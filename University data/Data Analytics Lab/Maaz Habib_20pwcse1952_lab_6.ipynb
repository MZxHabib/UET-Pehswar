```python
df.replace("?" ,"NaN", inplace=True)

#inplace=True.....applied to same dataframe
```

```python
df["normalized-losses"]= df["normalized-losses"].astype("float64")
df["bore"]= df["bore"].astype("float64")
df["stroke"]= df["stroke"].astype("float64")
df["peak-rpm"]= df["peak-rpm"].astype ("float64")
df["horsepower"]=df["horsepower"].astype("float64")

df["price"]= df["price"].astype("float64")
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
mean_normalizedlosses=df["normalized-losses"].mean()
mean_bore= df["bore"].mean()
mean_stroke= df["stroke"].mean()
mean_horsepower= df["horsepower"].mean()
mean_peakrpm=df["peak-rpm"].mean()
```

```python
#df.replace("?" ,"NaN", inplace=True)

df["normalized-losses"].replace(np.nan, mean_normalizedlosses, inplace=True)
df["bore"].replace(np.nan , mean_bore, inplace=True)
df["stroke"].replace(np.nan ,mean_stroke, inplace=True)
df["horsepower"].replace(np.nan ,mean_horsepower, inplace=True)
df["peak-rpm"].replace(np.nan , mean_peakrpm, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.dropna(subset=["price"], inplace=True)
```
